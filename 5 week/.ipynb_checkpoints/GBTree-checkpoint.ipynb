```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
from math import exp
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import log_loss
```

```python
data = pd.read_csv('gbm-data.csv')
data
```

```python
X = data[list(data.columns)[1:]]
Y = data['Activity']
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def fitnmark(Xtrain, Xtest, Ytrain, Ytest, rate):
    marks_test = list()
    marks_train = list()
    classifier = GradientBoostingClassifier(n_estimators=250, verbose=False, random_state=241, learning_rate=rate)
    classifier.fit(Xtrain, Ytrain)
    for i in classifier.staged_decision_function(Xtest):
        marks_test.append(log_loss(Ytest, sigmoid(i)))
    for i in classifier.staged_decision_function(Xtrain):
        marks_train.append(log_loss(Ytrain, sigmoid(i)))
    return marks_test, marks_train
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.8, random_state=241)
rates = [1, 0.5, 0.3, 0.2, 0.1]
```

```python
for i in rates:
    print "learning rate = {0}".format(i)
    scores_test, scores_train = fitnmark(X_train, X_test, y_train, y_test, i)
    plt.figure()
    plt.plot(scores_test, label='test')
    plt.plot(scores_train, label='train')
    plt.legend(loc = 'upper right')
    plt.show()
    if i == 0.2:
        print 'On test min loss = {0} on {1} iter'.format(np.min(scores_test), np.argmin(scores_test))
    plt.close()
```
