```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
from tqdm import tqdm
from sklearn.metrics import r2_score
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import KFold
```

```python
data = pd.read_csv('abalone.csv')
data['Sex'] = data['Sex'].map(lambda x: 1 if x == 'M' else (-1 if x == 'F' else 0))
data
```

```python
X = data[list(data.columns)[:-1]]
Y = data['Rings']
```

```python
def fitnmark(X, Y, tree_num):
    classifier = RandomForestRegressor(random_state=1, n_estimators=tree_num)
    validator = KFold(n_splits=5, shuffle=True, random_state=1)
    marks = list()
    for train, test in validator.split(X):
        Xtrain, Xtest = [X.loc[i] for i in train], [X.loc[i] for i in test]
        Ytrain, Ytest = [Y[i] for i in train], [Y[i] for i in test]
        classifier.fit(Xtrain, Ytrain)
        mark = r2_score(Ytest, classifier.predict(Xtest))
        marks.append(mark)
    return np.mean(marks)
```

```python
trees_len = [i for i in range(1, 51)]
marks = list()
for i in tqdm(trees_len):
    marks.append(fitnmark(X, Y, i))
```

```python
plt.plot(trees_len, marks)
plt.plot([1, 50], [0.52, 0.52])
plt.show()
```
